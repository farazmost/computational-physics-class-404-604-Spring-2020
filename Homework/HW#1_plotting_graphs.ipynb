## Homework #1:
We want to graph the function below
\begin{equation}
f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x)/5+....sin(nx)/n)
\end{equation}
from -$\pi$ to $\pi$. Choose n=11, 21 and 31 and plot three curves in one figure.

```python
import numpy as np
import matplotlib.pyplot as plt

#we create a function with inputs n and x which could be a single value or array
#and output the appropriate function value
def f(n,x):

    f=np.pi/2    #initialize the first value
    
    for i in range(n+1):       #loop through the int values from 0 to n 
        if i%2 == 1:           #only accept odd values of n 
            f= f + (2 * np.sin((i)*x)/(i))      
    return f
    
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-np.pi, np.pi, 0.001)    #set our x value bounds
n = [11,21,31,100,1000]                       #set our wanted n values
print('It can be seen that as n approaches a larger and larger value, the function divereges')
plt.figure(figsize=(15,10))

for i in range(len(n)):
    y = f(n[i],x)
    plt.plot(x,y,label = 'n = '+str(n[i]))
plt.title('f(x)=pi/2+2*(sin(x)+sin(3x)/3+sin(5x/5)+....sin(nx/n))')

ax = plt.gca()
ax.axhline(0, color='black', lw=2)
ax.axvline(0, color='black', lw=2)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
